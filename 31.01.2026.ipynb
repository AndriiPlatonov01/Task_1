```python
# 9. Створіть словник, де ключі — назви тварин, а значення — середня тривалість життя. 
# Знайдіть тварину з найбільшою тривалістю життя.

animals1 = {}
number1 = int(input("Insrt th numbr:"))

for i in range(number1):
    name1 = input("Enter the beast:")
    number2 = int(input("enter the age:"))
    animals1.update({name1:number2})

print(animals1)

max_years = 0
oldest_animal = ""

for name, years in animals1.items():
    if years > max_years:
        max_years = years
        oldest_animal = name

if oldest_animal:
    print(oldest_animal)
    print(max_years)
```

```python
# Запитуємо кількість тварин
number1 = int(input("Скільки тварин ви хочете ввести? "))
animals = {}

# Цикл для заповнення словника
for i in range(number1):
    name1 = input("Введіть назву тварини: ")
    number2 = int(input(f"Скільки років живе {name1}? "))
    animals[name1] = number2  # додаємо у словник

# Змінні для пошуку максимуму
max_years = 0
oldest_animal = ""

# Цикл для пошуку найстарішої тварини
for name, years in animals.items():
    if years > max_years:
        max_years = years
        oldest_animal = name

print("\n--- Результат ---")
if oldest_animal:
    print(f"Найбільша тривалість життя у тварини: {oldest_animal}")
    print(f"Вона живе в середньому {max_years} років.")
else:
    print("Список порожній.")
```

```python
# # 21. . Створіть словник, де ключі — назви курсів валют, а значення — зміна курсу за день. 
# Знайдіть валюту з найбільшим зростанням. 

dollars = {}
count = int(input("How many currencies? "))

for i in range(count):
    name1 = input("Name:")
    number1 = int(input("enter the naumber:"))
    dollars.update({name1: number1})

list1 = list(dollars.values())
max_value = max(list1)
index_max = list1.index(max_value)
list2 = list(dollars.keys())
result1 = list2[index_max]

print(f"Найбільше зростання:{result1} - {max_value}")
```

```python
# 23. Створіть словник, що містить назви компаній та кількість працівників. Виведіть компанії з більш ніж 500 працівниками.

comp = {}
count = int(input("How many comp:? "))

for i in range(count):
    name1 = input("Name:")
    number1 = int(input("enter the number:"))
    comp.update({name1: number1})

print(comp)

for comp, quat in comp.items():
    if quat > 500:
        print(quat)
```
